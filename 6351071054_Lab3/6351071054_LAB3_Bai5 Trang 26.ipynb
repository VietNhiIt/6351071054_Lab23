```python
from tensorflow.keras.datasets import mnist
import numpy as np
print('Nguyễn Viết Ái Nhi - 6351071054')
# Load MNIST
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Flatten ảnh 28x28 thành vector 784 chiều
X = train_images.reshape(-1, 28*28).astype('float32')

# Chuẩn hóa về [0,1]
X /= 255.0

y_true = train_labels  # Nhãn thật

from sklearn.cluster import KMeans

# Số cụm = số lớp MNIST = 10
kmeans = KMeans(n_clusters=10, random_state=42)

# Huấn luyện k-Means
kmeans.fit(X)

# Lấy nhãn cluster của từng ảnh
y_pred = kmeans.labels_

from sklearn.metrics import adjusted_rand_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Độ tương đồng giữa cluster và nhãn thật
ari = adjusted_rand_score(y_true, y_pred)
print(f"Adjusted Rand Index (ARI): {ari:.4f}")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix: Nhãn thật vs Cluster")
plt.xlabel("Cluster Label")
plt.ylabel("True Label")
plt.show()

```
