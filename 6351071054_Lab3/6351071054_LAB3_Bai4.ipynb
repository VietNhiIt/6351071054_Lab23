```python
from sklearn.datasets import load_iris
import numpy as np
print('Nguyễn Viết Ái Nhi - 6351071054')
# Đọc dữ liệu Iris
iris = load_iris()

# Lấy data và target
X = iris.data          # 4 thuộc tính
y = iris.target        # 3 lớp: 0, 1, 2

print("Tên các thuộc tính:", iris.feature_names)
print("Tên các lớp:", iris.target_names)
print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)
from sklearn.preprocessing import MinMaxScaler

# Chuẩn hóa về [0,1]
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

print("\n5 dòng đầu sau khi chuẩn hóa:")
print(X_normalized[:5])
```

**câu c**

```python
from keras.models import Sequential
from keras.layers import Dense
print('Nguyễn Viết Ái Nhi - 6351071054')
def create_model():
    model = Sequential()
    # Hidden layer 1
    model.add(Dense(10, activation='relu', input_shape=(4,)))
    # Hidden layer 2
    model.add(Dense(20, activation='relu'))
    # Output layer
    model.add(Dense(3, activation='softmax'))

    # Biên dịch mô hình
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

**Câu d**

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Nguyễn Viết Ái Nhi - 6351071054')
# Giả sử X_normalized, y đã được chuẩn hóa ở bước b
X_data = X_normalized
y_data = y

kfold = KFold(n_splits=10, shuffle=True, random_state=42)

models = []
accuracy_per_fold = []
precision_per_fold = []
recall_per_fold = []
f1_per_fold = []

fold_no = 1

for train_index, val_index in kfold.split(X_data):
    print(f"\n===== Fold {fold_no} =====")

    X_train_fold = X_data[train_index]
    y_train_fold = y_data[train_index]
    X_val_fold   = X_data[val_index]
    y_val_fold   = y_data[val_index]

    # Tạo model mới
    model = create_model()

    # Huấn luyện
    model.fit(
        X_train_fold, y_train_fold,
        epochs=100,
        batch_size=1,
        verbose=1,
        validation_data=(X_val_fold, y_val_fold)
    )

    # Dự đoán
    y_pred_prob = model.predict(X_val_fold)
    y_pred = np.argmax(y_pred_prob, axis=1)

    # Tính các metrics
    acc = accuracy_score(y_val_fold, y_pred)
    prec = precision_score(y_val_fold, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_val_fold, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_val_fold, y_pred, average='weighted', zero_division=0)

    print(f"Fold {fold_no} - Accuracy: {acc:.4f} - Precision: {prec:.4f} - Recall: {rec:.4f} - F1-score: {f1:.4f}")

    # Lưu kết quả
    models.append(model)
    accuracy_per_fold.append(acc)
    precision_per_fold.append(prec)
    recall_per_fold.append(rec)
    f1_per_fold.append(f1)

    # Confusion matrix
    cm = confusion_matrix(y_val_fold, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
    plt.title(f'Confusion Matrix Fold {fold_no}')
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

    fold_no += 1

# Metrics trung bình sau 10 folds
print("\n===== Metrics trung bình sau 10 folds =====")
print(f"Average Accuracy: {np.mean(accuracy_per_fold):.4f}")
print(f"Average Precision: {np.mean(precision_per_fold):.4f}")
print(f"Average Recall: {np.mean(recall_per_fold):.4f}")
print(f"Average F1-score: {np.mean(f1_per_fold):.4f}")
```
