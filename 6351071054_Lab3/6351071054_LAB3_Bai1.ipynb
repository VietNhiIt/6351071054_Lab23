```python
from sklearn.datasets import load_iris
import pandas as pd

# Đọc bộ dữ liệu Iris
iris = load_iris()
```

**câu b**

```python
print('Nguyễn Viết Ái Nhi - 6351071054')
# -----------------------------
# 1. In ra tên các thuộc tính và tên các lớp
# -----------------------------
print("Tên các thuộc tính (feature names):")
print(iris.feature_names)

print("\nTên các lớp (target names):")
print(iris.target_names)

# -----------------------------
# 2. In ra số lượng mẫu và số lượng thuộc tính
# -----------------------------
print("\nKích thước dữ liệu (data.shape):", iris.data.shape)
print("Kích thước nhãn (target.shape):", iris.target.shape)

# -----------------------------
# 3. In ra 5 mẫu dữ liệu đầu tiên
# -----------------------------
print("\n5 mẫu dữ liệu đầu tiên:")
print(iris.data[:5])
print("5 nhãn đầu tiên:")
print(iris.target[:5])

# -----------------------------
# 4. Thống kê dữ liệu bằng describe()
# -----------------------------
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print("\nThống kê mô tả của bộ dữ liệu:")
print(df.describe())
```
