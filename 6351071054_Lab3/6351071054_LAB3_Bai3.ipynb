```python
from keras.datasets import mnist

# Đọc dữ liệu MNIST
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
print("Kích thước tập train (images):", train_images.shape)
print("Kích thước tập train (labels):", train_labels.shape)

print("Kích thước tập test (images):", test_images.shape)
print("Kích thước tập test (labels):", test_labels.shape)
# Chuẩn hóa dữ liệu về [0,1]
train_images = train_images.astype("float32") / 255.0
test_images  = test_images.astype("float32") / 255.0

# Chuyển ảnh 28x28 thành vector 784 chiều
train_images = train_images.reshape((60000, 28*28))
test_images = test_images.reshape((10000, 28*28))

```

**câu d**

```python
from keras.models import Sequential
from keras.layers import Dense
print('Nguyễn Viết Ái Nhi - 6351071054')
def create_model():
    model = Sequential()

    # Hidden layer
    model.add(Dense(128, activation='relu', input_shape=(784,)))

    # Output layer
    model.add(Dense(10, activation='softmax'))

    # Compile model
    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

**câu e**

```python
from sklearn.model_selection import KFold
import numpy as np
print('Nguyễn Viết Ái Nhi - 6351071054')
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

models = []                 # Lưu mô hình sau mỗi fold
accuracy_per_fold = []      # Lưu accuracy mỗi fold
loss_per_fold = []          # Lưu loss mỗi fold

fold_no = 1

for train_index, val_index in kfold.split(train_images):
    print(f"\n===== Fold {fold_no} =====")

    # Tạo model mới từ hàm create_model()
    model = create_model()

    # Chia dữ liệu train/validation theo fold
    X_train_fold = train_images[train_index]
    y_train_fold = train_labels[train_index]
    X_val_fold   = train_images[val_index]
    y_val_fold   = train_labels[val_index]

    # Huấn luyện mô hình
    history = model.fit(
        X_train_fold, y_train_fold,
        epochs=5,
        batch_size=32,
        validation_data=(X_val_fold, y_val_fold),
        verbose=0   # 0 để không in quá nhiều
    )

    # Đánh giá mô hình trên validation fold
    loss, accuracy = model.evaluate(X_val_fold, y_val_fold, verbose=0)

    print(f"Fold {fold_no} - Loss: {loss:.4f} - Accuracy: {accuracy:.4f}")

    # Lưu mô hình
    models.append(model)

    # Lưu loss + accuracy
    loss_per_fold.append(loss)
    accuracy_per_fold.append(accuracy)

    fold_no += 1

# In kết quả chung
print("\n===== Kết quả trung bình sau 10 folds =====")
print("Average Accuracy:", np.mean(accuracy_per_fold))
print("Average Loss:", np.mean(loss_per_fold))
```
